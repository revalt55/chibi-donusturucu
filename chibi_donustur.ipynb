```python
from google.colab import files\nfrom PIL import Image\nimport torch\nfrom diffusers import StableDiffusionPipeline\nimport io
```

```python
pipe = StableDiffusionPipeline.from_pretrained("lambdalabs/pastel-mix-anime-1.0", torch_dtype=torch.float16)\npipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")
```

```python
uploaded = files.upload()\nfor name in uploaded:\n    img = Image.open(io.BytesIO(uploaded[name])).convert("RGB")\n    img = img.resize((512,512))
```

```python
prompt = "chibi style, cute character, pastel colors, soft anime features"\nimage = pipe(prompt=prompt, init_image=img, strength=0.75, guidance_scale=7.5, num_inference_steps=30).images[0]\nimage.save("chibi.png")\ndisplay(image)
```
